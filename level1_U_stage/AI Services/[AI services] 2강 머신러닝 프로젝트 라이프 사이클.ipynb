## 1. 머신러닝 프로젝트 플로우

### 1-1.문제 정의

특정 현상을 파악해 현상에 있는 문제를 정의하는 것, 구체적으로 명확이 무엇을 해야 하는지?를 파악해야하지

* 예시 : 사람들을 행복하게 하고 싶다.
    * 어떤 사람들? 행복의 기준은? 돈이냐 정신적이냐. 어떻게 할려구?
    
문제 해결을 위해선 구체적으로 정의를 해야하는데 어떻게 하면 좋냐면..

* 현상 파악
* 목적, 문제 정의 : 문제를 쪼개서 구체적으로, 왜 이런 문제가 있고 어떻게 해결해야할지 생각해보면
* 프로젝트 설계
* Action 행동
* 추가 원인 분석, 생각하지 못한 원인이 있나?

레스토랑 식당의 손님이 줄었다. -> 왜 줄었지? 물어보니 메뉴가 너무 많다더라. -> 그럼 이런것들은 어떻게 해결 할 수 있지

닭잡는데 소잡는칼 쓰지말자. 데이터로 할 수 있는 일을 찾아보고 꼭 알고리즘 접근이 아닌 단순하고 쉬운 방법으로도 제시 할 수 있어야 한다. 시간과 금액이 문제!

#### 머신러닝 프로젝트 과정
문제 정의를 하고 어떤 Metric으로 평가를 할지, 데이터 수집하고 레이블 정의하고 개발을 하고 마지막으로 모델이 왜 틀렸는지 에러 분석하기. 잘 돌다가 갑자기 안되면 왜 안되는지 파악해야하고, 재학습하고 metric 수정하고..


#### 프로젝트 설계
결국 최대한 구체적으로 설계 하는 것이 좋음
* 당장 할 수 있는것, 할 수는 없지만 고려해야할 사항들 정리

#### 머신러닝이 좋은 경우
1. 패턴이 있어야 함, 패턴이 없는 복권과 같은 사례는 어렵지
* 데이터가 없다면? 직접 수집하거나, 룰베이스 알고리즘으로 데이터 수집계획을 만들자.

2. 반복 : 사람이 반복적으로 행동해야하는 경우

#### 머신러닝이 사용되면 안좋은 경우
1. 비윤리적 문제
2. 간단히 해결될 수 있는
3. 좋은 데이터를 얻기 어렵고
4. 한번의 예측결과가 치명적인 결과를 발생할 경우
5. 시스템이 내리는 모든 결정을 설명 가능해야 하는 경우
6. 가성비 안나오는 경우

프로젝트의 목표

* Objectives : 목적을 달성하기 위한 세부 단계 목표(구체적인 목적)

    * 참여에 따른 게시물 랭킹 선정 : 사용자가 클릭할 가능성이 있는 게시물 추천
